```python
sales_category = data.groupby("Category")["Sales"].sum()

sales_category.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.show()
```

```python
region_sales = data.groupby("Region")["Sales"].sum()

region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.show()
```

```python
top_products = data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_products.plot(kind="bar")

plt.title("Top 10 Products")
plt.show()
```

```python
data["Order Date"] = pd.to_datetime(data["Order Date"])
```

```python
data["Month"] = data["Order Date"].dt.month
```

```python
monthly_sales = data.groupby("Month")["Sales"].sum()

monthly_sales.plot()

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(data.corr(numeric_only=True), annot=True)
plt.show()
```

```python
1. Technology category generated the highest sales.
2. Western region produced the highest revenue.
3. Sales peak in November and December.
4. Few products contribute to majority of sales.
```
